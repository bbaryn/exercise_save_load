### <a name='a2'></a> Za≈Çadowanie danych

```python
def fetch_financial_data(company='AMZN'):
    """
    This function fetch stock market quotations.
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data('META')
df.head()
```

### <a name='a3'></a> Zapis/Odczyt danych z pliku CSV

```python
df.to_csv('meta.csv')
```

```python
df_nov = df[(df.index.month == 11) & (df.index.year == 2019)]
df_nov
```

```python
df_nov.to_csv('meta_nov.csv')
```

```python
new_df = pd.read_csv('meta_nov.csv', index_col=0)
new_df
```

```python
df_nov.to_excel('meta_nov.xlsx')
```
